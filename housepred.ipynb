```python
df = df.drop(columns='LotFrontage')
```

```python
df.shape
```

```python
# Example of median imputation
df['MasVnrArea'].fillna(df['MasVnrArea'].median(), inplace=True)
```

```python
# Example of imputation with a new category
for col in ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']:
    df[col].fillna('Missing', inplace=True)
```

```python
# Example of mode imputation
df['Electrical'].fillna(df['Electrical'].mode()[0], inplace=True)
```

```python
# Example of imputation with mode for categorical features
for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']:
    df[col].fillna('No Garage', inplace=True)

# Impute numerical feature
df['GarageYrBlt'].fillna(df['GarageYrBlt'].median(), inplace=True)
```

```python
null_columns = df.isnull().sum()

# Filter out columns with null values
null_columns = null_columns[null_columns > 0]

# Print the column names and their respective null value counts
print(null_columns)
```

```python
df.head()
```

```python
low_cardinality_col = [col for col in df.columns if df[col].dtype == object and df[col].nunique() < 10]
high_cardinality_col = [col for col in df.columns if df[col].dtype == object and df[col].nunique() >= 10]
df_numerical = df.select_dtypes(include=['number'])
num_col = df_numerical.columns
```

```python
X = pd.concat([df[low_cardinality_col], df[high_cardinality_col], df[num_col]], axis=1)
```

```python
X.head()
```

```python
X.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_col = low_cardinality_col + high_cardinality_col
for col in cat_col:
 X[col] = le.fit_transform(X[col])
```

```python
X.isnull().sum()
```

```python
y = X['SalePrice']
X = X.drop(columns=['SalePrice'])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_cv, y_train, y_cv = train_test_split(X, y, test_size=0.4,
random_state=1)
```

```python
X_cv, X_test, y_cv, y_test = train_test_split(X_cv, y_cv, test_size=0.2, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
dtr = DecisionTreeRegressor()
rfr = RandomForestRegressor()
dtr.fit(X_train, y_train)
dtr_prediction = dtr.predict(X_cv)
rfr.fit(X_train, y_train)
rfr_prediction = rfr.predict(X_cv)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
dtr_error = mean_absolute_error(y_cv, dtr_prediction)
dtr_error_percentage = mean_absolute_percentage_error(y_cv, dtr_prediction)
dtr_error, dtr_error_percentage*100
```

```python
rfr_error = mean_absolute_error(y_cv, rfr_prediction)
rfr_error_percentage = mean_absolute_percentage_error(y_cv,
rfr_prediction)
rfr_error, rfr_error_percentage*100
```

```python
rfr_prediction = rfr.predict(X_test)
dtr_prediction = dtr.predict(X_test)
```

```python
dtr_error = mean_absolute_error(y_test, dtr_prediction)
dtr_error_percentage = mean_absolute_percentage_error(y_test, dtr_prediction)
dtr_error, dtr_error_percentage*100
```

```python
rfr_error = mean_absolute_error(y_test, rfr_prediction)
rfr_error_percentage = mean_absolute_percentage_error(y_test,
rfr_prediction)
rfr_error, rfr_error_percentage*100
```
